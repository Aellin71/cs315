```python
import numpy as np
from math import isqrt
```

```python

inString = input("Input the encrypted text")

dimensions = isqrt(len(inString))

matrix = np.array ( [] )
for x in inString:
  matrix = np.append (matrix, x)
newMatrix = matrix.reshape (dimensions, dimensions)

print(newMatrix)
```
